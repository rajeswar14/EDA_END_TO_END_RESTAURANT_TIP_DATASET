# Plot's for Continuous Data

**1.univariate (single variable)**
 - Histogram
 - Boxplot
 
**2.Bivariate (plot between two variables)**
- scatter plot 
- line plot
- join plot
- violin plot 

**3.Multivariate (more than 2 variables)**
- pair plot 
- Heatmap

# Histogram/Distribution Plot

```python
sns.histplot(df["total_bill"],bins=10,kde=True)#kernal density estimation
plt.show()
```

```python
sns.kdeplot(df['total_bill'])
plt.show()
```

# Box Plot

```python
sns.boxplot(x=df["tip"])
plt.show()
```

```python
sns.boxplot(x=df["total_bill"])
plt.show()
```

# scatter plot

```python
sns.scatterplot(x=df["total_bill"],y=df['tip'])
plt.show()
```

```python
sns.scatterplot(x=df["total_bill"],y=df["tip"],hue=df["sex"])
plt.show()
```

```python
sns.relplot(x=df["total_bill"],y=df["tip"],hue=df["sex"])
plt.show()
```

# Line Plot

```python
df['sno'] = pd.DataFrame(np.arange(1,245))
df
```

```python
sns.relplot(x = 'sno', y = 'tip',data = df,kind="line")
plt.show
```

```python
df.drop("sno",axis=1,inplace=True)
```

# JoinPlot 

```python
sns.jointplot(x='total_bill',y='tip',data=df)
plt.show()
```

# Violin Plot

```python
sns.violinplot(x='total_bill',y='sex',data=df)
plt.show()
```

# Pair Plot - Multiple continuous variables

```python
sns.pairplot(df,vars=continuous)
plt.show()
```

# Heatmap

```python
c_m = df[continuous].corr()
c_m
```

```python
sns.heatmap(c_m,annot=True)
plt.show()
```

# CountPlot

```python
df['smoker'].unique()
```

```python
df['smoker'].value_counts()
```

```python
sns.countplot(x='smoker',data=df)
plt.show()
```

```python
sns.countplot(y=df["smoker"])
plt.show()
```

# Barplot

```python
df.groupby("sex")["total_bill"].mean()
```

```python
sns.catplot(x='sex',y='total_bill',data=df,kind="bar")
plt.show()
```

```python
df.groupby("smoker")["total_bill"].mean()
```

```python
sns.catplot(x='smoker',y='total_bill',data = df,kind="bar")
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=df,hue='size',kind="bar")
plt.show
```

# Box Plot

```python
sns.catplot(x="day",y="total_bill",data=df, kind="box")
plt.show()
```
